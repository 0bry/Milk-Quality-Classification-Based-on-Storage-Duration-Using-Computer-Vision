# Training Model

```python
from ultralytics import YOLO

model = YOLO('yolo11n-cls.pt')

results = model.train(
    data='/content/milk4-1',
    epochs=10,
    imgsz=640,
    batch=32,
    augment=False,
    hsv_h=0.0,
    hsv_s=0.0,
    hsv_v=0.0,
    degrees=0.0,
    translate=0.0,
    scale=0.0,
    shear=0.0,
    perspective=0.0,
    flipud=0.0,
    fliplr=0.0,
    mosaic=0.0,
    mixup=0.0,
    copy_paste=0.0,
    auto_augment=None,
    erasing=0.0,
)
```
